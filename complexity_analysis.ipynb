## Time Complexity

Compare execution time on two systems: old system (less computation power) and new system (high computation power).

Same program with same data size takes 10 sec on old system and 1 sec on new system. *This doesn't mean, new system has better time complexity than old system*, as can be seen in below example.

$\text{Time Complexity } != \text{Time Taken}$

```python
import matplotlib.pyplot as plt
import math
```

```python
# x axis value
x = [10, 100, 200, 5000, 10000]

# y axis value
y = [10, 100, 200, 5100, 9000]

# plotting the points
plt.plot(x, y)

# naming the x axis
plt.xlabel('size')

# naming the y axis
plt.ylabel('time')

plt.title('old system')

plt.show()
```

```python
# x axis value
x = [10, 100, 200, 5000, 10000]

# y axis value
y = [1, 10, 20, 500, 1000]

# plotting the points
plt.plot(x, y)

# naming the x axis
plt.xlabel('size')

# naming the y axis
plt.ylabel('time')

plt.title('new system')

plt.show()
```

From above graphs, we can see the time complexity is linear in both the case, even though the time consumed on old system is larger than new system for same data size. Only the slope will be different.

**Time Complexity:** Function that gives us the relationship about how the time will grow as the input grows.

**Why Time complexity?**

Comparing time complexity of linear search $O(n)$ and binary search $O(log n)$

```python
# x axis value
x = [2000, 4000, 6000, 8000, 10000]

# y axis value
y = [5, 10, 15, 20, 25]

# plotting the points
plt.plot(x, y)

# naming the x axis
plt.xlabel('size')

# naming the y axis
plt.ylabel('time')

plt.title('Linear Search Time Complexity')

plt.show()
```

```python
# x axis values
x = [2000, 4000, 6000, 8000, 10000]

# y axis values
y = []
for value in x:
    y.append(math.log10(value))

# plotting the points
plt.plot(x, y)

# naming the x axis
plt.xlabel('size')

# naming the y axis
plt.ylabel('time')

plt.title('Binary Search Time Complexity')

plt.show()
```

As can be seen from above two graphs, as the size of data increases, the gap between time consumption between linear search and binary search increases, with binary search consuming less time.

This is the reason we need time complexity analysis, as is tells us which algorithm performs better. In our example, Binary search is better algorithm for searching than linear search.

```python
x = [2000, 4000, 6000, 8000, 10000]

linear_y = [1, 2, 3, 4, 5]
plt.plot(x, linear_y, label='linear')


log_y = []
for value in x:
    log_y.append(math.log10(value))
plt.plot(x, log_y, label='log')


constant_y = [2, 2, 2, 2, 2]
plt.plot(x, constant_y, label='constant')

plt.xlabel('size')
plt.ylabel('time')
plt.title('Multiple Time Complexity')

plt.legend()

plt.show()
```

From above graph it is evident, when data size grows: $O(1) < O(log \text{ n}) < O(n)$

**What needs to be considered when calculating complexity?**

- Always focus on worst case complexity.

- Always look at complexity for large or infinite data.

- Ignore constants.
    - ![Varying Linear Complexities](assets/images/time_complexity_1.png)
    
    - Even though the value of actual time is different, they are growing linearly.
    
    - We are not worried about the time taken. We focus on the relationship, how the time grows when input grows.

    - This is the reason for ignoring all the constants while computing complexity.

- Always ignore less dominating terms.
    - e.g. $O(n^{3} + log(n))$

    - \# of records = 1 million

    - time = $(1000000^{3} + log(1000000))$ = $(1000000^{3} + 6)$

    - as compared to $1000000^{3}$, $6$ is very small i.e. insignificant. Hence, ignore it.


**Big-Oh Notation**

<ins>Definition:</ins>
> The complexity (the graph) is the upper bound.
>
> In other words, the complexity or the graph or the relationship cannot exceed the upper limit.
>
> E.g.
> - Time Complexity: $O(n^{3})$
>
> - It might be possible that the algorithm gets solved in $O(n)$ or $O(log(n))$ or $O(n log(n))$ or $O(n^{2})$. But it will never exceed $O(n^{3})$.

<ins>Mathematical perspective:</ins>
> $f(n) = O(g(n))$
>
> $\lim_{n \to \infty} f(n)/g(n) < {\infty}$
> 
> E.g.
> - $g(n) = O(n^{3})$
>
> - $f(n) = O(6n^{3} + 3n + 5)$
>
> - $\lim_{n \to \infty} (6n^{3} + 3n + 5)/n^{3}$  
= $\lim_{n \to \infty} (6 + (3 / n^{2}) + (5 / n^{3}))$  
= $6 + (3 / \infty) + (5 / \infty)$  
= $6 + 0 + 0$  
= $6 < \infty$
